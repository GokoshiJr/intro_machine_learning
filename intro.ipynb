```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('./data/diabetes.csv') # read in your CSV

data.head() # get a sneak preview of the dataset

data["Age"] # age feature
data["Age"].shape # cantidad de registros filas x columnas
data[["Age", "Insulin"]] # age and insulin feature
data.dtypes # muestra el tipo de representacion de los datos

# print("Número original de datapoints\n", data.shape) # filas x columnas
# elimina cualquier valor faltante y remplazarlos hacia la variable de datos
# data = data.dropna()

# Comprueba la nueva forma (shape)
# print("Número actualizado de datapoints\n", data.shape) # filas x columnas

# Seleccionar las filas donde el valor es mayor a 0.
data = data[data["BloodPressure"] > 0]
print("Patients with blood pressure\n", data.shape) # Rows x Columns

# los histogramas son generalmente buenos para obtener una distribucion de un feature
# en cambio para casos mas avanzados de multiples distribuciones,los graficos de dispersion
# extraordinariamente buenos para encontrar relaciones multidimensionales (entre dos features)

plt.rcParams['font.size'] = 15 
f = plt.figure(figsize=(8,4))
ax = f.add_subplot(1,1,1)
ax.scatter(data["BloodPressure"],data["Age"],alpha=0.5)
ax.set_xlabel("Blood Pressure")
ax.set_ylabel("Age")
ax.set_title("Age range of patients vs Blood Pressure")
f.tight_layout()
# ax.set_xlim([20,130])
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('./data/diabetes.csv') # read in your CSV

# histograma para obtener la distribucion del rango de edad de los pacientes
f = plt.figure(figsize=(8,6))
ax = f.add_subplot(1,1,1)
data["Age"].hist(ax=ax, edgecolor="black", linewidth=2)
ax.set_title("Age range of patients")
ax.set_ylim([0,500])
ax.set_xlabel("Age")
ax.set_ylabel("Count")
f.tight_layout()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('./data/diabetes.csv') # read in your CSV
data = data.dropna() # Elimina cualquier valor faltante y remplazarlos hacia la variable de los datos
# Histograma cantidad de diabeticos en el dataset
f = plt.figure(figsize=(8,4)) # Crear una figura y ajusta su tamaño.
ax = f.add_subplot(1,1,1) # Crear un subplot o subtrama - al ser una sola figura es 1,1,1.
data["Outcome"].hist(ax=ax, edgecolor='black', linewidth=2) # Gráfica tus datos usando 'hist'. Pasa el objeto 'ax' a Pandas. Agrega un borde negro con un groso de 2.
ax.set_xlim([-0.5, 1.5]) # Establece los límites en el eje x.
ax.set_xticks([0, 1]) # Establece la frecuencia de tick. Tenemos 0 y 1 que corresponden a Sí y No respectivamente.
ax.set_xticklabels(["N", "Y"]) # Etiquetar xtick labels
ax.set_title("Diabetes Y/N?") # Crea el título.
ax.set_xlabel("Answer") # Establece la etiqueta del eje X.
ax.set_ylabel("Count") # Establece la etiqueta del eje Y.
ax.set_ylim([0, 510]) # Establece los límites superior/inferior del eje y.
# Hace que las cosas sean bonitas, no es necesario, pero se ajusta al tamaño de la figura.
f.tight_layout()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('./data/diabetes.csv') # read in your CSV

# histograma para obtener la distribucion del rango de edad de los pacientes
bin_values = [20,25,30,35,40,85]
f = plt.figure(figsize=(8,6))
ax = f.add_subplot(1,1,1)
data["Age"].hist(ax=ax, bins=bin_values, edgecolor="black", linewidth=2) # bins=5
ax.set_title("Age range of patients")
ax.set_ylim([0,500])
ax.set_xlabel("Age")
ax.set_ylabel("Count")
f.tight_layout()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('./data/diabetes.csv') # read in your CSV
diabetic = data[data["Outcome"] == 1]
notdiabetic = data[data["Outcome"] == 0]

plt.rcParams['font.size'] = 15 
f = plt.figure(figsize=(8,8))
ax = f.add_subplot(2,1,1)
ax.scatter(diabetic["BloodPressure"], diabetic["Age"], alpha=0.25, c='r')
ax.set_title("Diabetic: Age range of patients v. Blood Pressure")
ax.set_ylabel("Age")
ax.set_xlabel("Blood Pressure")

ax = f.add_subplot(2,1,2)
ax.scatter(notdiabetic["BloodPressure"], notdiabetic["Age"], alpha=0.25, c='b')
ax.set_title("Diabetic: Blood Pressure v. Age range of patients")
ax.set_ylabel("Blood Pressure")
ax.set_xlabel("Age")

f.tight_layout()
```
