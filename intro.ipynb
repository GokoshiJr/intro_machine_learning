```python
import pandas as pd
import matplotlib.pyplot as plt
# read in your CSV
data = pd.read_csv('./diabetes.csv')

# get a sneak preview of the dataset
data.head()
data["Age"]
data["Age"].shape # cantidad de registros
data[["Age", "Insulin"]]

data.dtypes # muestra el tipo de representacion de los datos

# los histogramas son generalmente buenos para obtener una distribucion
# en cambio para casos mas avanzados de multiples distribuciones,los graficos de dispersion
# extraordinariamente buenos para encontrar relaciones multidimensionales
ax = data.plot.scatter(x="BloodPressure", y="Age", alpha=0.5)
ax.set_title("Age range of patients vs Blood Pressure")
```

```python
f = plt.figure(figsize=(8,4))
ax = f.add_subplot(1,1,1)
data["Age"].hist(ax=ax, edgecolor="black", linewidth=2)
ax.set_title("Age range of patients")
ax.set_ylim([0,510])
ax.set_xlabel("Age")
ax.set_ylabel("Count")
f.tight_layout()
```
